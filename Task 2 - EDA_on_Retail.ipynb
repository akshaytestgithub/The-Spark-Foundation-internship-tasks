```python
count_of_furniture = data.Category.value_counts()['Furniture']
count_of_furniture
```

```python
count_of_office_supplies = data.Category.value_counts()['Office Supplies']
count_of_office_supplies
```

```python
count_of_technology = data.Category.value_counts()['Technology']
count_of_technology
```

```python
category_names = ['Technology', 'Office Supplies', 'Furniture']
custom_colours = ['#F08080', '#9ACD32', '#40E0D0']
sizes = [count_of_technology, count_of_office_supplies, count_of_furniture]
plt.figure(figsize = (5,5), dpi = 100)
plt.pie(sizes, labels = category_names, textprops = {'fontsize' : 12}, colors = custom_colours, autopct = '%0.2f%%', pctdistance = 0.4, labeldistance = 1.1, startangle = 90, explode = [0.05, 0.05, 0.05])
center_circle = plt.Circle((0,0), radius = 0.25, fc = 'white')
plt.gca().add_artist(center_circle)
plt.show()
```
